```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

breast_cancer_df = pd.read_csv('breast-cancer.csv')
covid_df = pd.read_csv('covid.csv')
```

```python
# Get counts for both breast cancer and covid
breast_cancer_phases_df = breast_cancer_df['Phases'].value_counts(dropna=False)
covid_phases_df = covid_df['Phases'].value_counts(dropna=False)

# Create a bar chart for breast cancer
plt.figure(figsize=(8, 6))
bars = breast_cancer_phases_df.plot(kind='bar', color='hotpink')

# Add numbers on top of bars
for i, value in enumerate(breast_cancer_phases_df.values):
        bars.text(i, value, value, ha='center', va='bottom')

plt.xlabel('Phases')
plt.ylabel('Count')
plt.title('Distribution of Phases in the Breast Cancer DataFrame')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the bar chart for breast cancer
plt.show()

#################################################################################

# Create a bar chart for covid
plt.figure(figsize=(8, 6))
bars = covid_phases_df.plot(kind='bar', color='mediumorchid')

# Add numbers on top of bars
for i, value in enumerate(covid_phases_df.values):
        bars.text(i, value, value, ha='center', va='bottom')

plt.xlabel('Phases')
plt.ylabel('Count')
plt.title('Distribution of Phases in the Covid DataFrame')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the bar chart for covid
plt.show()
```

```python
# Count the frequency of each intervention type in each dataset
breast_cancer_phases_df = breast_cancer_df['Phases'].value_counts(dropna=False)
covid_phases_df = covid_df['Phases'].value_counts(dropna=False)

# Create a new DataFrame with these counts
df = pd.DataFrame({'Breast Cancer': breast_cancer_phases_df, 'COVID': covid_phases_df})

# Fill NaN values with 0
df = df.fillna(0)

# Plot the DataFrame using pandas' built-in plot function
df.plot(kind='bar', figsize=(12, 6), color=['hotpink', 'mediumorchid'])

plt.title('Comparison between Phase Counts in Breast Cancer and COVID Trials')
plt.ylabel('Number of Occurrences')
plt.xlabel('Phase')

plt.show()
```
