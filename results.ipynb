```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

breast_cancer_df = pd.read_csv('breast-cancer.csv')
covid_df = pd.read_csv('covid.csv')
```

```python
# Get the number of 'Yes' and 'No' answers for breast cancer
n_yes = breast_cancer_df['Study Results'].value_counts()['YES']
n_no = breast_cancer_df['Study Results'].value_counts()['NO']

# Create a pie chart for breast cancer
plt.figure(figsize=(6, 6))
plt.pie([n_yes, n_no], labels=['Yes', 'No'], autopct="%1.1f%%")
plt.title('Distribution of Results in the Breast Cancer DataFrame')
plt.show()

####################################################################

# Get the number of 'Yes' and 'No' answers for covid
n_yes = covid_df['Study Results'].value_counts()['YES']
n_no = covid_df['Study Results'].value_counts()['NO']

# Create a pie chart for breast cancer
plt.figure(figsize=(6, 6))
plt.pie([n_yes, n_no], labels=['Yes', 'No'], autopct="%1.1f%%")
plt.title('Distribution of Results in the Covid DataFrame')
plt.show()
```
