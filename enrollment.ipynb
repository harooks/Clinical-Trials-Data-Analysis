```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

breast_cancer_df = pd.read_csv('breast-cancer.csv')
covid_df = pd.read_csv('covid.csv')
```

```python
# Calculate total enrollment for each DataFrame
total_enrollment_bc = breast_cancer_df['Enrollment'].sum()
total_enrollment_c = covid_df['Enrollment'].sum()

# Create labels for the bar chart
labels = ['Breast Cancer', 'Covid']

# Create the bar chart
plt.figure(figsize=(6, 6))
bars = plt.bar(labels, [total_enrollment_bc, total_enrollment_c], color=['hotpink', 'mediumorchid'])
plt.xlabel('DataFrames')
plt.ylabel('Total Enrollment (Millions)')
plt.title('Comparison of Total Enrollment Between Breast Cancer and Covid Clinical Trials')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Modify y-axis ticks and labels to show thousands
plt.yticks([0, 1000000, 2000000, 3000000, 4000000], [0, 1, 2, 3, 4])

# Add labels on top of each bar
for bar, value in zip(bars, [total_enrollment_bc, total_enrollment_c]):
    x = bar.get_x() + bar.get_width() / 2  # Center the text horizontally
    y = value + 0.1  # Offset the text slightly above the bar
    plt.text(x, y, str(value), ha='center', va='bottom', fontsize=12)

plt.tight_layout()

# Display the bar chart
plt.show()
```
